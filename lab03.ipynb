## Introduccion

The choice for this laboratory is CIFAR-10. CIFAR-10 is a labeled image dataset consisting of 60,000 color images divided into 10 classes. Each image has spatial structure and multiple color channels, making it suitable for convolutional neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print("Training samples:", x_train.shape[0])
print("Test samples:", x_test.shape[0])
print("Image shape:", x_train.shape[1:])
```

```python
unique, counts = np.unique(y_train, return_counts=True)

plt.bar(range(10), counts)
plt.xlabel("Class")
plt.ylabel("Number of samples")
plt.title("Class distribution in CIFAR-10")
plt.show()
```

The dataset is balanced, with approximately the same number of samples per class.
### Examples

```python
class_names = [
    'airplane', 'automobile', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

plt.figure(figsize=(10,6))
for i in range(10):
    idx = np.where(y_train == i)[0][0]
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[idx])
    plt.title(class_names[i])
    plt.axis('off')
plt.show()
```

Visual inspection shows that objects may appear at different positions and scales within the images, motivating the use of convolutional layers to capture local patterns and translation-invariant features.
